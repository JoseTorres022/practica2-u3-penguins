Las columnas son:

- species: penguin species (Chinstrap, Adélie, or Gentoo)
- culmen_length_mm: culmen length (mm)
- culmen_depth_mm: culmen depth (mm)
- flipper_length_mm: flipper length (mm)
- body_mass_g: body mass (g)
- island: island name (Dream, Torgersen, or Biscoe) in the Palmer -   - Archipelago (Antarctica)
- sex: penguin sex

1. species (especie): Esta variable representa la especie de pingüino a la que pertenece cada entrada en el conjunto de datos. Hay tres especies de pingüinos incluidas en este conjunto de datos: Chinstrap, Adélie y Gentoo. Cada entrada estará etiquetada con una de estas especies.

2. culmen_length_mm (longitud del culmen en mm): El culmen es el pico de un ave. Esta variable mide la longitud del culmen de cada pingüino en milímetros. La longitud del culmen es una característica importante para distinguir entre diferentes especies de pingüinos.

3. culmen_depth_mm (profundidad del culmen en mm): Esta variable mide la profundidad del culmen de cada pingüino en milímetros. Al igual que la longitud del culmen, la profundidad del culmen es una característica útil para identificar las especies de pingüinos.

4. flipper_length_mm (longitud de la aleta en mm): Mide la longitud de la aleta de cada pingüino en milímetros. La longitud de la aleta puede variar entre especies y es un indicador importante en la clasificación de los pingüinos.

5. body_mass_g (masa corporal en gramos): Esta variable indica la masa corporal de cada pingüino en gramos. La masa corporal es otra característica crítica para diferenciar entre especies y también puede proporcionar información sobre la salud y el estado de los pingüinos.

6. island (isla): Representa el nombre de la isla en la que se encuentra cada pingüino. Las tres islas mencionadas son Dream, Torgersen y Biscoe, que forman parte del Archipiélago Palmer en la Antártida. La ubicación geográfica puede tener un impacto en las características de los pingüinos.

7. sex (sexo): Indica el sexo de cada pingüino, ya sea macho o hembra.

**Observaciones:**

- El culmen es el pico
- flipper  (aletas) =Las alas de los pingüinos se llaman aletas. Son planos, delgados y anchos con una forma larga y cónica y una punta roma y redondeada.

## 4.1 Análisis de cada variable de manera individual
Esto nos permitira entender las caracteristicas generales de cada variable del dataset.

```python
# Primero que ndad, veremos las variables categoricas.
# Son del tipo: Dtype=object()
data.info()
```

```python
# Veamos las variables categoricas en graficos de barras.
columnas_categoricas = ['species', 'island', 'sex']

# Graficas de barras de conteo
fig, ax = plt.subplots(nrows=len(columnas_categoricas),
                       ncols=1, figsize=(8, 20))
fig.subplots_adjust(hspace=0.5)

for i, col in enumerate(columnas_categoricas):
    sns.countplot(x=col, data=data, ax=ax[i])
    ax[i].set_title(col)
    ax[i].set_xticklabels(ax[i].get_xticklabels(), rotation=30)
```

```python
# Volvamos a ver las variables del dataset
print("Variables del dataset:")
data.info()

# Ahora, extraemos las variables estadisticas descriptivas basicas.
print("\nVariables estadisticas descriptivas basicas:")
data.describe()
```

```python
# Es mejor graficasr estas variables estadisticas.

# Dibujaremos hisogramas
columnas_numericas = ['bill_length_mm', 'bill_depth_mm',
                      'flipper_length_mm', 'body_mass_g', 'year']

# Cuantos graficos se mostraran
fig, ax = plt.subplots(nrows=5, ncols=1, figsize=(10, 30))
fig.subplots_adjust(hspace=0.5)

for i, col in enumerate(columnas_numericas):
    if col == 'bill_length_mm':
        nbins = 10
    else:
        nbins = 50
    sns.histplot(x=col, data=data, ax=ax[i], bins=nbins, kde=True)
    ax[i].set_title(col)
```
