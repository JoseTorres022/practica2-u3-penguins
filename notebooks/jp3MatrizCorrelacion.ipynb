```python
# Importando liberias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#import numpy as np
```

```python
# Lectura del dataset
ruta = "../results/dataset_penguins_clean.csv"
data = pd.read_csv(ruta)
#print(data)

data.columns #ver las comlumnas del CSV
data.describe() #ver variables numericas
#data.isna() #ver datos fatantes
#data.isna().sum() #contar la cantidade datos faltantes#
```

```python
#Filtrando comlumnas numericas
filtro = ['bill_length_mm','bill_depth_mm',
          'flipper_length_mm', 'body_mass_g', 'year']
data=data[filtro]

#Mostrando e nuevo DF
data
```

```python
#Creando la mtriz de correlacion
matriz_correlacion=data.corr().round(2)
matriz_correlacion
```

```python
# Diagrama de Calor
plt.figure(figsize=(10,8))
sns.heatmap(data=matriz_correlacion, #pasandoles los datos CSV con el nuevo DF
            annot=True, #Mostrar los datos dentos de los cuadros
            annot_kws={'size': 12}, #TamaÃ±o de los datos en los cuadros
            linecolor='white', #Color del borde de los cuadros
            linewidths=0.75) # Grosor del borde de los cuadros
plt.show()
```
